## k-최근접 이웃 회귀
* 전에 했던 분류와 다르게 물고기의 길이를 x로 물고기 무게를 y로 정의
* 물고기의 길이를 입력했을때 무게를 예측하는 예제

### 데이터 준비

```python
import numpy as np
```

```python
perch_length = np.array(
    [8.4, 13.7, 15.0, 16.2, 17.4, 18.0, 18.7, 19.0, 19.6, 20.0,
     21.0, 21.0, 21.0, 21.3, 22.0, 22.0, 22.0, 22.0, 22.0, 22.5,
     22.5, 22.7, 23.0, 23.5, 24.0, 24.0, 24.6, 25.0, 25.6, 26.5,
     27.3, 27.5, 27.5, 27.5, 28.0, 28.7, 30.0, 32.8, 34.5, 35.0,
     36.5, 36.0, 37.0, 37.0, 39.0, 39.0, 39.0, 40.0, 40.0, 40.0,
     40.0, 42.0, 43.0, 43.0, 43.5, 44.0]
     )
perch_weight = np.array(
    [5.9, 32.0, 40.0, 51.5, 70.0, 100.0, 78.0, 80.0, 85.0, 85.0,
     110.0, 115.0, 125.0, 130.0, 120.0, 120.0, 130.0, 135.0, 110.0,
     130.0, 150.0, 145.0, 150.0, 170.0, 225.0, 145.0, 188.0, 180.0,
     197.0, 218.0, 300.0, 260.0, 265.0, 250.0, 250.0, 300.0, 320.0,
     514.0, 556.0, 840.0, 685.0, 700.0, 700.0, 690.0, 900.0, 650.0,
     820.0, 850.0, 900.0, 1015.0, 820.0, 1100.0, 1000.0, 1100.0,
     1000.0, 1000.0]
     )
```

### 데이터셋 분리

```python
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target=train_test_split(perch_length, perch_weight, random_state=42)
```

```python
# 사이킷런은 훈련 셋을 2차원 배열이어야함!!!
train_input=train_input.reshape(-1, 1)
test_input=test_input.reshape(-1, 1)
```

### 결정계수(R^2) 구하기

```python
from sklearn.neighbors import KNeighborsRegressor

knr=KNeighborsRegressor()

knr.fit(train_input, train_target)
```

```python
print(knr.score(test_input, test_target))
```

```python
# 잘 점수가 나왔나 한번 계산해보자
from sklearn.metrics import mean_absolute_error

test_prediction=knr.predict(test_input) #테스트 세트에 대한 예측을 만들자

mae=mean_absolute_error(test_target, test_prediction) #테스트 세트에 대한 평균 절댓값 오차를 계산
print(mae)

#오차가 19정도 됨
```

### 과대적합 과소적합

```python
print(knr.score(train_input, train_target)) #train에 대한 점수가 test에 대한 점수보다 낮게 나옴!! 과소적합!!!
```

```python
# knn에서 모델을 복잡하게 만드려면 k의 수를 줄여라!! 그럼 국지적인 패턴에 민감해질것임

knr.n_neighbors=3
knr.fit(train_input, train_target)
print(knr.score(train_input, train_target))
print(knr.score(test_input,test_target))

#테스트셋에 대한 점수가 더 낮게 나옴! 그리고 얼마 차이도 안남! 구웃
```

## 정리
* knn은 분류와 회귀 둘 다 가능
  * 분류의 경우는 주변 이웃들의 정보를 보고 분류
  * 회귀의 경우는 주변 이웃들의 정보의 평균을 낸 점수로 결정

* 데이터 준비 -> 전처리 -> train/test 분리 -> 모델 shape 2차원으로 변경 -> 모델에 넣어 훈련 -> 결정계수 계산 -> 과소적합, 과대적합 판단 -> 만약 과소적합이면 k의 개수를 줄여서 계산 -> 마무리~
